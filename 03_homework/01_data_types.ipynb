### 1. What types are involved in the following expressions? What data type will each expression evaluate to?

`8 * 2.5`

```python
# Your code here
Out = 8 * 2.5

print(Out)

print(f"This is: {type(Out)}")
```

`9 / 2`

```python
# Your code here
Out = 9 / 2

print(Out)

print(f"This is: {type(Out)}")
```

`9 // -2`

```python
# Your code here
Out = 9 // -2

print(Out)

print(f"This is: {type(Out)}")
```

`1.5 * 2 >= 7 - 3`

```python
# Your code here
Out = 1.5 * 2 >= 7 - 3

print(Out)

print(f"This is: {type(Out)}")
```
